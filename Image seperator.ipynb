```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
img=cv2.imread("canola-field-nature-background-19710623.jpg",)
```

```python
i_img=img
```

```python
i_img=i_img.reshape(-1,3)
```

```python
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
img_=img.reshape(-1,3)
```

```python
print(img_.shape)
```

```python
from sklearn.cluster import KMeans
```

```python
diff_colours=4
km=KMeans(n_clusters=diff_colours)
```

```python
km.fit(img_)
```

```python
labels=km.labels_
```

```python
print(len(km.labels_))
```

```python
centers=km.cluster_centers_
```

```python
colours=np.array(centers,dtype='uint8')
```

```python
print(colours)
```

```python
new_image=np.zeros(img_.shape,dtype='uint8')
```

```python
for i in range(new_image.shape[0]):
    if labels[i]==1 or labels[i]==3 :
        new_image[i]=img_[i]
    else:
        new_image[i]=i_img[i]
    
```

```python
new_image=new_image.reshape(img.shape)
```

```python
plt.imshow(new_image)
```
